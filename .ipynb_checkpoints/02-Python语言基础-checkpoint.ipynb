## 2.2 Python 语言基础

**缩进**

Python语言使用缩进（tab 或者 空格）来组织代码，而不是像其他语言（比如说C++，Jva或者R语言）一样使用大括号来表示代码块。

```python
# 这是一个例子
num = 10

while num <= 15:
    print(num)
    num += 1
```

一个冒号 : 代表一个缩进代码块的开始，同一个代码块中的所有代码必须保持相同的缩进。

**变量和参数传递**

在Python语言中，对一个变量进行赋值的时候，用户创建了一个指向等号右边对象的引用。

```python
a = [1, 2, 3]
```

```python
b = a
```

此时a和b指向的是同一个对象。

```python
id(a)
```

```python
id(b)
```

可以发现a和b两个变量的内存地址是相同的。

```python
a = [2, 3, 4]  # 我们来修改a列表的内容
```

```python
id(b)
```

可以发现，当我们修改了a列表中的某一个元素的时候，我们实际上开辟了一个新的内存空间，将原本的列表拷贝到这个内存空间中，然后修改了其中的元素。

## 2.3 控制流

**if-elif_else**

if是最广为人知的控制流语句。它检查一个条件，如果为True，就执行后面的语句：

```python
x = -10

if x < 0:
    print("It's negative")
```

if后面可以跟一个或多个elif，所有条件都是False时，还可以添加一个else：

```python
if x < 0:
    print("It's negative")
elif x == 0:
    print('Equal to zero')
elif 0 < x < 5:
    print('Positive but smaller than 5')
else:
    print('Positive and larger than or equal to 5')
```

**for**

for循环是在一个集合（列表或元组）中进行迭代，或者就是一个迭代器。for循环的标准语法是：

```python
sequence = [1, 2, None, 4, None, 5]
total = 0
for value in sequence:
    if value is None:
        continue
    total += value
```

可以用break跳出for循环。下面的代码将各元素相加，直到遇到5：

```python
sequence = [1, 2, 0, 4, 6, 5, 2, 1]
total_until_5 = 0
for value in sequence:
    if value == 5:
        break
    total_until_5 += value
```

break只中断for循环的最内层，其余的for循环仍会运行：

```python
In [121]: for i in range(4):
   .....:     for j in range(4):
   .....:         if j > i:
   .....:             break
   .....:         print((i, j))
```

**While循环**

while循环指定了条件和代码，当条件为False或用break退出循环，代码才会退出：

```python
x = 256
total = 0
while x > 0:
    if total > 500:
        break
    total += x
    x = x // 2
```

```python
if x < 0:
    print('negative!')
elif x == 0:
    # TODO: put something smart here
    pass
else:
    print('positive!')
```
